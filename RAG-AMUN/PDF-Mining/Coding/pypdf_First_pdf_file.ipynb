```python
# get the pdfreader to read the pdf file the we got.
from pypdf import PdfReader, PdfWriter
```

```python
# Create a reader object to read the file
reader = PdfReader('Test (1).pdf')
```

```python
print(len(reader.pages)) # the length of the pages in the pdf file
```

```python
pages = {}
for i in range(len(reader.pages)):
    page = reader.pages[i]
    pages[f'page_{i}'] = page.extract_text()
    pages[f'page_{i}'] = pages[f'page_{i}'].replace('\n', ' ')
```

```python
pages['page_0']
```

```python
for page in range(len(reader.pages)):
    print(f'page {page} is :', pages[f'page_{page}'])
```

```python
def format_paragraphs(input_text):
    # Split the input text into lines
    lines_1 = input_text.split('اﻟﻌﻣدة:')
    output_text = lines_1[0]

    for line in range(1, len(lines_1)):
        output_text += '\n اﻟﻌﻣدة:' + lines_1[line]

    lines_2 = output_text.split('ﺟﻌﯾدى:')

    output_text_final = lines_2[0]

    for line in range(1, len(lines_2)):
        output_text_final += '\n ﺟﻌﯾدى:' + lines_2[line]

    return output_text_final

# Example usage
output_text = format_paragraphs(pages['page_0'])
print(output_text)
```

```python
new_pages = {}

for page in range(len(reader.pages)):
    output_text = format_paragraphs(pages[f'page_{page}'])
    new_pages[f'page_{page}'] = output_text
    print(f'page {page} is: \n \n', new_pages[f'page_{page}'], '\n')
```

```python
writer = PdfWriter()

for page in range(len(reader.pages)):
    writer.add_page(reader.pages[page])

with open("PyPDF2-Test-1.pdf", "wb") as fp:
    writer.write(fp)
```
