```python
# get the pdfreader to read the pdf file the we got.
from pypdf import PdfReader
```

```python
# Create a reader object to read the file
reader = PdfReader('Test (2).pdf')
```

```python
print(len(reader.pages)) # the length of the pages in the pdf file
```

```python
pages = {}
for i in range(len(reader.pages)):
    page = reader.pages[i]
    pages[f'page_{i}'] = page.extract_text()
```

```python
from PyPDF2 import PdfReader, PdfWriter
```

```python
reader = PdfReader('Test (2).pdf')
```

```python
pages = {}
for i in range(len(reader.pages)):
    page = reader.pages[i]
    pages[f'page_{i}'] = page.extract_text()
```

```python
for page in range(len(reader.pages)):
    print(f'page {page} is : \n \n', pages[f'page_{page}'], '\n')
```

```python
writer = PdfWriter()

for page in range(len(reader.pages)):
    writer.add_page(reader.pages[page])

with open("PyPDF2-Test-2.pdf", "wb") as fp:
    writer.write(fp)
```

```python
reader = PdfReader('Test (3).pdf')
```

```python
len(reader.pages)
```

```python
pages = {}
for i in range(len(reader.pages)):
    page = reader.pages[i]
    pages[f'page_{i}'] = page.extract_text()
    pages[f'page_{i}'] = pages[f'page_{i}'].replace('\n', ' ')
```

```python
pages['page_0']
```

```python
for page in range(len(reader.pages)):
    print(f'page {page} is :', pages[f'page_{page}'])
```

```python
def format_paragraphs(input_text):
    # Split the input text into lines
    lines_1 = input_text.split('اﻟﺬﺋﺐ  :')
    output_text = lines_1[0]

    for line in range(1, len(lines_1)):
        output_text += '\n اﻟﺬﺋﺐ  :' + lines_1[line]

    lines_2 = output_text.split('اﻟﺜﻌﻠﺐ:')

    output_text_2 = lines_2[0]

    for line in range(1, len(lines_2)):
        output_text_2 += '\n الثعلب:' + lines_2[line]

    lines_3 = output_text_2.split('اﻷرﻧﺐ  :')

    output_text_3 = lines_3[0]

    for line in range(1, len(lines_3)):
        output_text_3 += '\n اﻷرﻧﺐ  :' + lines_3[line]

    lines_4 = output_text_3.split('اﻷﺳﺪ  :')

    output_text_4 = lines_4[0]

    for line in range(1, len(lines_4)):
        output_text_4 += '\n اﻷﺳﺪ  :' + lines_4[line]

    lines_5 = output_text_4.split('اﻟﺜﻌﻠﺐ  :')

    output_text_5 = lines_5[0]

    for line in range(1, len(lines_5)):
        output_text_5 += '\n اﻟﺜﻌﻠﺐ  :' + lines_5[line]

    return output_text_5

# Example usage
output_text = format_paragraphs(pages['page_0'])
print(output_text)
```

```python
for page in range(len(reader.pages)):
    print(f'page {page} is : \n \n', format_paragraphs(pages[f'page_{page}']), '\n')
```
