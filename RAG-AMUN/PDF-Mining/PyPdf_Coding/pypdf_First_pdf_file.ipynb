# Dealing with the first File - Test (1).pdf

## Read The File Using PdfReader

```python
# Create a reader object to read the file
reader = PdfReader('Test (1).pdf')
```

```python
print(len(reader.pages)) # the length of the pages in the pdf file
```

## Save the pages into a dictionary to process them

```python
pages = {}
for i in range(len(reader.pages)):
    page = reader.pages[i]
    pages[f'page_{i}'] = page.extract_text()
    pages[f'page_{i}'] = pages[f'page_{i}'].replace('\n', ' ')
    pages[f'page_{i}'] = ''.join(['(' if char == ')' else ')' if char == '(' else char for char in pages[f'page_{i}']])
```

```python
pages['page_0']
```

## Print the pages

```python
for page in range(len(reader.pages)):
    print(f'page {page} is :', pages[f'page_{page}'])
```

## Try to solve some problems with first file

```python
def format_paragraphs(input_text):
    # Split the input text into lines
    lines_1 = input_text.split('اﻟﻌﻣدة:')
    output_text = lines_1[0]

    for line in range(1, len(lines_1)):
        output_text += '\n اﻟﻌﻣدة:' + lines_1[line]

    lines_2 = output_text.split('ﺟﻌﯾدى:')

    output_text_2 = lines_2[0]

    for line in range(1, len(lines_2)):
        output_text_2 += '\n ﺟﻌﯾدى:' + lines_2[line]

    lines_3 = output_text_2.split('ﺳﻌدﯾﺔ:')

    output_text_3 = lines_3[0]

    for line in range(1, len(lines_3)):
        output_text_3 += '\n ﺳﻌدﯾﺔ:' + lines_3[line]

    lines_4 = output_text_3.split('ﺳﻌدﯾﺔ– ﺻﺑرﯾﺔ:')

    output_text_4 = lines_4[0]

    for line in range(1, len(lines_4)):
        output_text_4 += '\n ﺳﻌدﯾﺔ– ﺻﺑرﯾﺔ:' + lines_4[line]

    lines_5 = output_text_4.split('ﺻﺑرﯾﺔ:')

    output_text_5 = lines_5[0]

    for line in range(1, len(lines_5)):
        output_text_5 += '\n ﺻﺑرﯾﺔ:' + lines_5[line]

    lines_6 = output_text_5.split('أﺣﻣد:')

    output_text_6 = lines_6[0]

    for line in range(1, len(lines_6)):
        output_text_6 += '\n أﺣﻣد:' + lines_6[line]

    lines_final = output_text_6.split('اﻟﺟﻣﯾﻊ:')

    output_text_final = lines_final[0]

    for line in range(1, len(lines_final)):
        output_text_final += '\n اﻟﺟﻣﯾﻊ:' + lines_final[line]

    return output_text_final

# Example usage
output_text = format_paragraphs(pages['page_0'])
print(output_text)
```

## Print the text after extraction and processing

```python
new_pages = {}

for page in range(len(reader.pages)):
    output_text = format_paragraphs(pages[f'page_{page}'])
    new_pages[f'page_{page}'] = output_text
    print(f'page {page} is: \n \n', new_pages[f'page_{page}'], '\n')
```

## Save the text into a txt file with name - Test (1).txt

```python
f = open('Test (1).txt', 'w')
for page in range(len(pages)):
    output_text = format_paragraphs(pages[f'page_{page}'])
    f.write(output_text)
f.close()
```

```python
def format_paragraphs(input_text):
    # Split the input text into lines
    lines_1 = input_text.split('اﻟﺬﺋﺐ  :')
    output_text = lines_1[0]

    for line in range(1, len(lines_1)):
        output_text += '\n اﻟﺬﺋﺐ  :' + lines_1[line]

    lines_2 = output_text.split('اﻟﺜﻌﻠﺐ:')

    output_text_2 = lines_2[0]

    for line in range(1, len(lines_2)):
        output_text_2 += '\n الثعلب:' + lines_2[line]

    lines_3 = output_text_2.split('اﻷرﻧﺐ  :')

    output_text_3 = lines_3[0]

    for line in range(1, len(lines_3)):
        output_text_3 += '\n اﻷرﻧﺐ  :' + lines_3[line]

    lines_4 = output_text_3.split('اﻷﺳﺪ  :')

    output_text_4 = lines_4[0]

    for line in range(1, len(lines_4)):
        output_text_4 += '\n اﻷﺳﺪ  :' + lines_4[line]

    lines_5 = output_text_4.split('اﻟﺜﻌﻠﺐ  :')

    output_text_5 = lines_5[0]

    for line in range(1, len(lines_5)):
        output_text_5 += '\n اﻟﺜﻌﻠﺐ  :' + lines_5[line]

    return output_text_5
```
